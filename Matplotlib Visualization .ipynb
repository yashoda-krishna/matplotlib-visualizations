```python
import matplotlib.pyplot as plt

x=["a","b","c","d"]
y=[2,4,6,8]
z=[1,2,3,1]

plt.xlabel("A",fontsize=15)
plt.ylabel("X",fontsize=15)
plt.title("BAR",fontsize=15)

c=["y","g","r"]
d=["g","r","y"]

plt.bar(x,y,width=0.5,color=c,align="edge",label="rushil")

plt.bar(x,z,width=0.5,color="b",align="edge",label="ruhii")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["a","b","c","d"]
y=[2,4,6,8]
z=[1,2,3,1]

p=np.arange(len(x))

width=0.3
p1=[j+width for j in p]

plt.xlabel("A",fontsize=15)
plt.ylabel("X",fontsize=15)
plt.title("BAR",fontsize=15)

c=["y","g","r"]
d=["g","r","y"]

plt.bar(p,y,width,color=c,align="edge",label="rushil")

plt.bar(p1,z,width,color="b",align="edge",label="ruhii")

plt.xticks(p+width,x,rotation=20)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,3,1,1]

c=["y","b","y","b"]
#jetla point etla color apva padse

size=[100,200,300,400]

plt.scatter(x,y,color=c,s=size,alpha=1,marker="*",edgecolor="black",linewidth=1)

#alpha == transparency
plt.title("Scatter",fontsize=20)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
  
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,3,5,10]
z=[1,1,1,1] 

color=[20,30,50,100]
#Seamap no between 0 to 100

#jetla point etla color apva padse

size=[100,200,300,400]

plt.scatter(x,y,c=color,s=size,cmap="Accent")

plt.scatter(x,z,s=size,cmap="Viridis")

x=plt.colorbar()
x.set_label("Colorbar")

#alpha == transparency
plt.title("Scatter",fontsize=20)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
  
```
